```python
X=df.iloc[:,:-1]
y=df["Species"]
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y= encoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
help(SVC)
```

```python
model = SVC(C=1)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.model_selection import KFold
kfold =KFold() 
n_folds =5
```

```python
parameters ={"C":[0.1,1,10,100,1000],"kernel":["poly","linear","rbf"]}
```

```python
from sklearn.model_selection import GridSearchCV
gridcv = GridSearchCV( model,parameters,cv=n_folds,scoring='accuracy',return_train_score=True)
gridcv.fit(X_train,y_train)
```

```python
print("best prams",gridcv.best_params_)
```

```python
print(gridcv.best_score_)
```

```python
result=pd.DataFrame(gridcv.cv_results_)
```

```python
result
```
